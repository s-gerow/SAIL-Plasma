```python
from threading import Event
import numpy
import pyvisa
```

```python
class ExperimentEvent(Event):
    def __init__(self,x="",y="",z=""):
        super().__init__()
        self.x = x
        self.y = y
        self.z = z
    def show(self):
        output = [self.x]
        return output
```

```python
triggered = ExperimentEvent()
```

```python
triggered.set()
```

```python
triggered.is_set()
```

```python
triggered.x=1
triggered.y=2
triggered.z=3
```

```python
frehcn = triggered.show()
print(frehcn)
```

```python
triggered.is_set()
```

```python
class newDevice(pyvisa.Resource):
    def __init__(self, rm, name):
        self.rm = rm
        self.name = name
        self.resource = None
    
    def open_device(self):
        device_list = rm.list_resources()
        try:
            self.resource = rm.open_resource(self.name)
        except pyvisa.errors.VisaIOError:
            print(f"Unable to connect to resource, provide a valid device name: {device_list}")

    def isOpen(self):
        if self.resource.resource_name in self.rm.list_opened_resources():
            return True
        else: return False

    def close(self):
        if self.isOpen():
            self.resource.close()
        else: return

```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


#t_trigger = 1749748821.804677
t_trigger = 1749822102.4767401
dmmseries = pd.read_csv('C:/Users/gerows/Python/SAIL-Plasma/dmm_timeseries.csv')
time = np.array(dmmseries.iloc[:,0].values)
voltages = np.array(dmmseries.iloc[:,1].values)
idx = np.argmin(np.abs(time-t_trigger))
voltage = voltages[idx]
voltage
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111)
plt.plot(time-time[0], voltages, label='Data')
plt.axvline(t_trigger-time[0], color='gray', linestyle='--', label='Discharge')
plt.legend()
plt.title("DMM Timeseries for Breakdown at 0.705 cm*Torr")
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
