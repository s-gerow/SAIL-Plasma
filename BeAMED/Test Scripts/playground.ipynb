```python
from threading import Event
import numpy
import pyvisa
```

```python
class ExperimentEvent(Event):
    def __init__(self,x="",y="",z=""):
        super().__init__()
        self.x = x
        self.y = y
        self.z = z
    def show(self):
        output = [self.x]
        return output
```

```python
triggered = ExperimentEvent()
```

```python
triggered.set()
```

```python
triggered.is_set()
```

```python
triggered.x=1
triggered.y=2
triggered.z=3
```

```python
frehcn = triggered.show()
print(frehcn)
```

```python
triggered.is_set()
```

```python
class newDevice(pyvisa.Resource):
    def __init__(self, rm, name):
        self.rm = rm
        self.name = name
        self.resource = None
    
    def open_device(self):
        device_list = rm.list_resources()
        try:
            self.resource = rm.open_resource(self.name)
        except pyvisa.errors.VisaIOError:
            print(f"Unable to connect to resource, provide a valid device name: {device_list}")

    def isOpen(self):
        if self.resource.resource_name in self.rm.list_opened_resources():
            return True
        else: return False

    def close(self):
        if self.isOpen():
            self.resource.close()
        else: return

```

```python
#from chatgpt
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

# 🔹 Sample data (non-smooth at knot)
np.random.seed(0)
x = np.linspace(0, 4, 50)
y = np.piecewise(x, [x < 2, x >= 2],
                 [lambda x: 2*x**2 - x + 1,
                  lambda x: -1.5*x**2 + 10*x - 10]) + np.random.normal(0, 2, size=x.shape)

# 🔹 Split the data at the knot
knot = 2.0
x_left = x[x <= knot]
y_left = y[x <= knot]
x_right = x[x > knot]
y_right = y[x > knot]

# 🔹 Degree of each polynomial
deg = 2

# 🔹 Objective function: sum of squared errors
def objective(coeffs):
    a = coeffs[:deg+1]
    b = coeffs[deg+1:]
    y_pred_left = np.polyval(a, x_left)
    y_pred_right = np.polyval(b, x_right)
    error = np.concatenate([(y_pred_left - y_left), (y_pred_right - y_right)])
    return np.sum(error**2)

# 🔹 Constraint functions
def c0_continuity(coeffs):
    a = coeffs[:deg+1]
    b = coeffs[deg+1:]
    return np.polyval(a, knot) - np.polyval(b, knot)

def c1_continuity(coeffs):
    a = coeffs[:deg+1]
    b = coeffs[deg+1:]
    a_deriv = np.polyder(a)
    b_deriv = np.polyder(b)
    return np.polyval(a_deriv, knot) - np.polyval(b_deriv, knot)

# 🔹 Initial guess: use separate polyfits
a0 = np.polyfit(x_left, y_left, deg)
b0 = np.polyfit(x_right, y_right, deg)
initial_guess = np.concatenate([a0, b0])

# 🔹 Run optimizer with constraints
constraints = [
    {"type": "eq", "fun": c0_continuity},   # C⁰ continuity
    {"type": "eq", "fun": c1_continuity},   # C¹ continuity
]

result = minimize(objective, initial_guess, constraints=constraints)

# 🔹 Extract solution
a_fit = result.x[:deg+1]
b_fit = result.x[deg+1:]

# 🔹 Plot result
x_fit = np.linspace(0, 4, 200)
y_fit = np.piecewise(x_fit, [x_fit <= knot, x_fit > knot],
                     [lambda x: np.polyval(a_fit, x),
                      lambda x: np.polyval(b_fit, x)])

plt.figure(figsize=(8, 5))
plt.plot(x, y, 'o', label='Data')
plt.plot(x_fit, y_fit, 'r-', label='Piecewise Poly Fit (C¹ continuous)')
plt.axvline(knot, color='gray', linestyle='--', label='Knot')
plt.legend()
plt.title("Piecewise Polynomial Fit with C¹ Continuity at x = {}".format(knot))
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.tight_layout()
plt.show()
```
