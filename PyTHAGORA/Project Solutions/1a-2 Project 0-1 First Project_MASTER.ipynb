#### **(P3)**

The math module provides useful functions as well as constant values such as pi. Lets use those. Write a script to find the cirumference of one end, the surface area, and volume of any cylinder using the radius and height. Then print these values. For the value of pi, you will need to use `math.pi`

*Hint: circumference is $2\times\pi\times r$, volume is $2\times\pi\times h\times r^2$, surface area is $2\times\pi\times r^2 + 2\times\pi\times r\times h$*

*The instructor solution uses r = 1 and h = 5 if you would like to compare your results, units are not necessarily important though we will revisit this problem in [3-6 Astropy](../Module3_DataTasks/3-6%20Astropy%20Lesson.ipynb) to test the `astropy.units` functionality.*

**P3 Solution**

>The circumference of a circle is given by the equation $c = 2\pi r$ or $c=\pi d$. The surface area of a cylinder is given by $A = 2\pi r^2 + \pi d h$ or $2\pi r^2 + c h$. The volume can be found with $V = \pi r^2 h$. Using these formulas, and $r = 1 \text{ and } h=5$, and `math.pi` we can determine the circumference, surface area, and volume of a cylinder. We can change the values of r and h to whatever we want and this will return the correct results.

````python
radius = 1
height = 5
pi = math.pi

circumference = 2*pi*radius
surface_area = 2*pi*radius**2 + circumference*height
volume = pi*radius**2*height

print(f"The circumference of a cylinder with radius {radius} and height {height} is {circumference},\n the surface area is {surface_area},\n and the volume is {volume}")
````

> The `print()` function above makes use of an f-string. An f-string is a more condensed way to write variable information into a string. I cover f-strings in more depth in [1c-3 While Loops](./1c-3%20While%20Loops.ipynb) 

#### **(P4)**

Given the equations $y = 7x$ and $x = \sqrt{y}$, find the points that the two lines intersect. This problem will be primarily algebra and less programming, but in remember what steps you take because in [3-3 Sympy](../Module3_DataTasks/3-3%20Sympy%20Lesson.ipynb) we will cover how to do symbolic algebra with Python and the package Sympy by revisiting this problem.



**(P4.1)** Solve each equation for y in terms of x (one is already done for you).

**P4.1 Solution**
>$x = \sqrt{y}$   
>$x^2 = \sqrt{y}^2$ - square each side of the equation to remove the radical   
>$y = x^2$ -this equation is solved for y in terms of x   

**(P4.2)** Set the two equations equal to each other and solve for x.

**P4.2 Solution**   
>$x^2 = 7x$ - replacing the y in one equation with the other equation results in this   
>$x^2 -7x = 0$ - by subtracting the right side x term to the left we can create a quadratic equation which can be solved using the quadratic formula: $x = \frac{-b \pm\sqrt{b^2-4ac}}{2a}$ and Python.


**(P4.3)** Find the solutions to the quadratic equation and their corresponding y-values.   
*Hint: the standard form of a quadratic equation can be written $ax^2 +bx +c = 0$*   
*Hint: the math module has a square root function*

**P4.3 Solution**

````python
a = 1 #a is the coefficient of the x^2 term, in this case 1
b = -7 #b is the coefficient of the x term, in this case -7
c = 0 #c is the number without an in it, in this case there is no third term so c is 0

#there are two possible x-values, one with a minus and one with a plus in the equation, we will denote these with x1 and x2
x1 = (-b + math.sqrt(b**2-(4*a*c)))/(2*a)
x2 = (-b - math.sqrt(b**2-(4*a*c)))/(2*a)
print(f"x1:{x1}\nx2:{x2}")
````

>Using these two x-values we can plug them into either equation from the beginning to find their coresponding y-values.   
>$y = 7(7) = 49$ and $y = (7)^2 = 49$ yeild the same result: a coordinate pair (7,49) as one of the intersection points.   
>$y = 7(0) = 0$ and $y=(0)^2$ also yeild the same result: a coordinate pair (0,0) as the other intersection point.
