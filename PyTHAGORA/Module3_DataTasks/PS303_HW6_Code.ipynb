```python
import matplotlib.pyplot as plt
from scipy.constants import h, c, k
from math import e
import numpy as np

fig, ax = plt.subplots()
for i in [2000, 3000, 4000, 5000, 6000]:
    T = i
    lambda_ = np.arange(1E-9, 5000E-9,1E-9)
    u_lambda = (8*pi*h*c*lambda_**-5)/((e**((h*c)/(k*T*lambda_)))-1)
    ax.plot(lambda_, u_lambda, label = f'u(lambda) at {T}K')
legend = ax.legend()
ax.set_xlabel('Wavelenth in nm')
ax.set_ylabel('u(lambda) [W/m^3]')
plt.show()
```
