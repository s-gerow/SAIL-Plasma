```python
fruits = ('apples', 'oranges', 'peaches', 'pears', 'mangos')

for fruit in fruits:
    print("*", fruit)
```

```python
more_fruits = list(fruits)
for fruit in ['strawberries', 'raspberries', 'kiwis']:
    more_fruits.append(fruit)
print(more_fruits)
```

(P1.3) Iterate through the list of fruits you created, asking the user where they would rank each fruit. Store the rank and fruit as a key-value pair in a dicitonary.

```python
fruit_rankings = {}
for fruit in more_fruits:
    rank = input(f"From 1 to {len(more_fruits)}, how would you rank {fruit}?")
    fruit_rankings.__setitem__(rank, fruit)
```

(P1.4) Print the fruit rankings in order.

```python
for rank in range(1,len(fruit_rankings)+1):
    print(rank, fruit_rankings[str(rank)])
```

# Interactive Phonebook
This problem will not need the PrimeHub
For some practice with while loops, *hint hint*, write a program which runs until the user quits using a designated exit phrase. Has an option to write in a new contact. And has an option to look at the current contacts list.

```python
#since I have not covered dictionaries as much, I have provided a dictionary example for you if you would like to use it.
#this exercise can be done without a dictionary just as easily.
contacts = [{'Name': 'Blank',
            'Number': '000-000-0000'}]
print(contacts[0]['Name'])
```

```python
while True:
  choice = input("Press 1 to add a contact. Press 2 to read the phonebook. Press e to exit")
  if choice == '1':
    name = input("Name:")
    number = input("Number:")
    contacts.append({'Name': name,
                    'Number': number})
  if choice == '2':
    for contact in contacts:
      print(contact)
  if choice == 'e':
    break
```

# Finding the true size of a matrix
In the Lists lesson I mentioned a bonus challenge; if you already did it, great, otherwise here it is.
Create a program which takes a nested list (matrix) input and returns the size in this format (x,y)

for example: (2x4) or (1,0)

```python
matrix = [[1,1,1,1],
          [1,1,1,1]]
rows = 0
col_counter = []
for row in matrix:
  rows += 1
  col_counter.append(len(row))
  if sum(col_counter)%len(col_counter) == 0:
    cols = int(sum(col_counter)/len(col_counter))
print(f"({rows},{cols})")
```
