##Booleans
In python there is a datatype called bool, for boolean. The only two values in python that carry this type are 'True' and 'False'. Booleans are a subclass of int meaning they can be used in the same way as numbers and mathmatical phrases. They are also a keyword which means they also cannot be used as a variable. Here's what I mean:

```python
True = 7 #Running this code results in a syntax error because true is a keyword and cannot be used as a variable.
```

```python
true = 7 #running this code is fine because python is case sensitive,
#it is advised that you do not use true as a variable to avoid confusion in your code
```

```python
fact = True #this statement can be run and the variable fact is known to be True
```

##Operators
The most important aspect of booleans is the boolean operators, these operators take a set of inputs and return either true or false. These operators will be imperative when learning to use if statements and while loops, which we will get to shortly.


###Not
The 'not' operator is the only 'single-input' operator that ever really matters. it returns the opposite of whatever the input is.

```python
fact = True
not fact #running this code returns false because that is the opposite of the input, true
```

Notice that only one thing was inputted into the operator, there are three other operators that take only one input but they largely are not useful simply due to their nature, a good resource if you would like to explore that more is [here](https://realpython.com/python-boolean/).

### And
The 'and' opperator takes two separate inputs and returns True if both of the inputs are true and otherwise returns false. Below is a conveinent truth table showing the output of the And operator based on different inputs.

```python
truth_table = [['Input A', 'Input B', 'Output'],['True', 'True', 'True'],['True', 'False', 'False'],['False', 'True', 'False'],['False', 'False', 'False']]
for row in truth_table: #This is a for loop, we will learn how they work next chapter, for now just run the code to view the truth table
    print('| {:^7} | {:^7} | {:^6} |'.format(*row))
```

```python
1+2==2 and 3+4==7 #In this example, the operator returns false because the first operation is not true.
```

###Comparison

The == does not mean to set equal to, it means to compare two values or expressions and determine if they are true. So 1+2==2 will return a value of false.



**<,<=,>,>=:**

### Or
The 'Or' operator takes two separate inputs and returns True if either of the inputs is true. Again, here is a truth table and example.

```python
truth_table = [['Input A', 'Input B', 'Output'],['True', 'True', 'True'],['True', 'False', 'True'],['False', 'True', 'True'],['False', 'False', 'False']]
for row in truth_table: #This is a for loop, we will learn how they work next chapter, for now just run the code to view the truth table
    print('| {:^7} | {:^7} | {:^6} |'.format(*row))
```

## If Statements
An if statement is the most basic control structure you can use, it works a lot like how it sounds: if this, then this.


Lets look at a way to apply booleans and operators within an if statement. First lets consider how we make decisions in our day to day lives.

Lets say you are trying to decide what to get for lunch on a certain day. What are some things you might think?

"What am I craving?"

"How hungry am I?"

"What have I eaten for lunch recently?"

If we take a look at some of these questions we can narrow them down to a simple "If this, then this".

"If I am craving fried chicken, then I should get Chick-fil-a"

"If I am very hungry, then I should order double fries"

"If I ate Chick-fil-a yesterday, then I shouldn't eat it again today"


This is how an "If Statement" works in Python, first it takes an input and then it compares it to some known value or expectation, then it does something. An if statement is the first control structure we will look at. If the phrase after the "if" in an if statement is True then Python will run the code contained within the statment, otherwise it does not. This provides a way for you to only run a set of code when a specific thing happens. Lets look at a basic example.

```python
x = 2
y = 5

if type(x) == int:
  print(x,"is a number")
if y >= 5:
  print(y,"is greater than or equal to 5")
```

Note the syntax of the if statement. First it starts with a lowercase "if", next is the item that is being checked. Usually this term is a variable; something that you don't know when writing the code, maybe a user input or the output of a function that you need to check the answer of. After the first value there is a boolean operator, followed by your constant condition that you are checking against, then a colon. The colon indicates that indented lines after this line are apart of the control structure designated by the if statement.


**Indentation Syntax** *It is important to understand that different control structures in python are all identified by line indents. The first line of a new structure, such as "if type(x) ==int:", provides python with the context for when to run the code contained within the structure. Python identifies the code that is contained in the structure by looking for line indents. If a line after a control structure is iniliazed is indented then that line will only be ran if the structure is ran. This is a very common error and is easy to miss.*

In the above example, the variable x is being checked to see if it is an integer type. If the variable x is an integer then the statement will run the rest of the code, if the variable is not an integer then it will not run that code.

```python
x = 'char'
y = 5

if type(x) == int:
  print(x,"is a number")
```

Notice that when you run this code there is no output, the variable x is not an integer type, it is a string type, so the if statement is not run.

### If-Else Statements
An extention of the if statement is the else clause. In an else-if combination, the code contained in the if statement will run if the comparison is true but if it is false then it will evaluate the code in the else statement. Let's look at an example. Say you are running a bar and need to know to turn away anyone under 21. You could check with a simple if-else statement.

```python
age = 19

if age >= 21:
  print("You can enter the bar")
else:
  print("You cannot enter the bar")
```

When the code is ran, the age given is not allowed to enter the bar, try changing the age to different numbers to see what happens.
Also, notice the indentations used. The if statement is not indented but the following line is. Then the else statement is not indented, that is to signify that the if statment has ended and a new control structure has started. You do not need a line of code to "end' the structure, simple unindent and continue to code.

### If-elif-else Statements
If statements have yet another extension that can be sused to make them more powerful: the elif statement. In many occasions you will need to test more than two conditions; say for instance you are selling movie tickets and on Tuesdays they are half-off for everyone, but also students and people over 65 get half off tickets every day, you want to test multiple conditions that could be true, this would be done with an if-elif-else statement.

```python
person = {'age': 21, 'student': True, 'day': 'Wednesday'} #Change some of these values and see if you can get different things to print
ticket_cost = 10

if person['age'] >= 65:
  print('65+ Ticket:  $', ticket_cost/2)
elif person['student'] == True:
  print('Student ticket:  $', ticket_cost/2)
elif person['day'] == 'Tuesday':
  print('Half-off-Tuesday Ticket:  $', ticket_cost/2)
else:
  print('Regular Ticket:  $', ticket_cost)
```

**Dictionaries** *The person variable above is a dictionary type, we did not discuess these in depth but they work like a list and a tuple except instead of being created with a bracket [] or parenthesis (), they are created with a curly bracket {}. Within the dictionary you can have any number of "key-value pairs" which include a string containing a "key", followed by a colon, and then a value which is attached to that key. Different key-value pairs are separated by commas. Dictionaries are indexed with a bracket and the the key which you want to find. In the above example there are three key-value pairs, 'age', 'student' and 'day'. Notice each of these have associated values which are a different type, age has an int value, student has a bool value, and day has a str value. This is to show that you can assign any type of value to a key in a variable. Dictionaries are quite useful in large scale datasets but are not as powerful as the Numpy library which we will explore more in a later chapter.*

In the if-elif-else example, only one of those values can be true. Once one of the if or elif values is found to be true then python will evaluate that and move on to later code. Try this by changing some of the values so that the person is both a student and the day is a Tuesday, note which code is executed.

Since only one value can be true in an if-elif chain, if you want multiple values to evaluate if they are true then you should use a chain of if statements, that way if every statement evaluates to true you will get an outtput for each true statement.
