# Tuples
Lists are great for an expanding dataset, when you want to allow a collection of data to expand over time and continue to collect new data. Sometimes, though, you want a list that does not change, Python has specfic lists which are unchangable, or immutable. These lists are called tuples.


##Creating and Indexing Tuples
Creating a tuple is done the same way as creating a list, except you use parenthesis instead of brackets.
Say for example you needed to reference the months of the year, but you don't want to accidentally write over your list of months, you could create a tuple, containing the name of each month.


```python
months = ('January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', ' November', 'December')
```

This tuple can be indexed the same as a list and each of the months can be pulled out and used in your program. Just like lists and strings though, remember that the first item in a tuple is the 0th item.

```python
print(months[0])
print(months[5])
```

Tuples can be nested just like lists and are indexed the same way.

```python
matrix = ((1,2),(3,4))
print(matrix[0][0])
print(matrix[0][1])
print(matrix[1][0])
print(matrix[1][1])
```

## Editing Tuples

Lets now try to rewrite the contents of the tuple.

```python
months[0] = 'December'
```

Notice that it says that a tuple does not support item assignment. This is because tuples are immutable. This will prevent you from accidentally changing a list that you want to remain constant. If you do want to change a tuple though, you can do so by re-assign a tuple value to the name of the variable.

```python
print(months)
months = ('jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec')
print(months)
```
