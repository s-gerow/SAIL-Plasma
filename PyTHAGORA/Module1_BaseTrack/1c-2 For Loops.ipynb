#More Control Stuctures
You have seen an if statement, "if this is true, do this"; lets look at another control structure which allows you to do something for some length of time: the for loop.

##The For Loop
Let's go back to lists for a moment; say you have a list of people you are inviting to a party, and you need to send each one an invite, you could write a line of code for each individual, but that would require you to copy and past each line over and over, changing only the name. Instead you can employ a for loop. Look at this example:

```python
invitees = ['mary','jane','john','monty']

for invitee in invitees:
  print(invitee.title(), 'is cordially invited to my birthday bash')
```

The for loop, as it is called a loop, will run the same code over and over until some criteria is met. In this case it says that for each item (invitee) in this list of invitees, print this message, so the loop will run over and over until it has printed a message for each person in the list. Invitee is a variable which changes each time the loop is ran, first it is Mary, then it is Jane, and so on until the list is complete.


### Types of sequences
A for loop iterates over any type of sequence; lists, tuples, dictionaries, sets, and strings are all types of sequences that could be iterated over. Here are some important cases to take note of.

#### Dictionaries
We have not specifically done any work with dictionaries but here is another example regarding their use. Say you are polling movie-goers at a movie theater. There are 3 movies available to watch and a full snack bar. You ask each person what movie they are seeing and what stnacks they bought. You then stored this information in a dictionary. A list of dictionaries was formed, where each dictionary represented a different persons response and the entire list represented all of the responses.

```python
poll_results = [{'name': 'susie', 'movie': 'How to Train your Dragon', 'snack': 'M&Ms'},
                {'name': 'john', 'movie': 'Shrek', 'snack': 'popcorn'},
                {'name': 'seth', 'movie': 'Interstellar', 'snack': 'skittles'},
                {'name': 'mary', 'movie': 'Shrek', 'snack': 'popcorn'}]
```

Now lets say you wanted to iterate over this list and return all of the movies that were watched by people today. You would do this with a for loop.

```python
movies = []
for response in poll_results:
  movies.append(response['movie'])
print(movies)
```

This for loop runs once for each 'response' in the list 'poll_results'. The variable 'response' changes each time the loop runs, starting with the first dictionary and ending with the last. The line inside the loop says that each time the loop is ran, it will append (add a value to a list) the 'movie' value to the list 'movies'. This is great but notice that there is a duplicate. Two people mentioned that they were seeing Shrek. If we had a list of 500 people then our movies list would be filled with duplicate movies. To get around this we can use the set() method. A set is a list with only unique items, so all duplicates will be removed.

```python
movies = []
for response in poll_results:
  movies.append(response['movie'])
print(set(movies))
```

#### Lists, Tuples, and Sets
The for loop is most used to loop over a set of data and retreive a single item to do something to. We saw an example of a loop iterating over a list to send out invites. Many applicatons of for loops will look like that just with different things being done to the value retrieved from the list.

As a final note, in "The Basics" chapter, I provided and example of a for loop that looked like this:

```python
last_name = 'Python'
length = len(last_name)
for i in range(0, length):
  print(last_name[i])
```

While this technically works it is more complicated than it needs to be. I used this example primarily to show datatypes and a breif introduction to indexing. Since in this case, the ith item of the list is being acted on, we can simply iterate over the list itself. Just like in the above example where the 'response' variable became the dictionary representing the response, by iterating over the list instead of the length, the 'i' variable will become the letter we want to print.

```python
for i in last_name:
  print(i)
```

```python
for i in 'Python': #this might make what is happening a bit more clear
  print(i)
```

As you can see, instead of 3 lines of code this is now 2 lines. In this small example it may not seem like it matters much, but in large programs, efficiency and clarity will become incredibly important.
