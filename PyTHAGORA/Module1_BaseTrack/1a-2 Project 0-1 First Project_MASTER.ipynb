#### **(P1)**

Consider the error message produced from running the following code, then re-write the cell to fix the error.
```
>>>print(hello world)
Cell In[2],   line 1
    print(hello world)
          ^
SyntaxError: invalid syntax. Perhaps you forgot a comma?
```

**P1 Solution**

````python
print(hello world)
````

>The error is due to "hello world" not being surrounded by quotes so Python does not know it is supposed to be printing a string. Instead it thinks that hello and world are different variables; if they were, they would need to be separated by a comma to be printed properly, which is why Python suggests using a comma to separate them.   
Because they are not actually two separate variables, when we make this change we will then get a NameError since hello and world are not defined.

````python
print(hello, world)
````

Both of these errors can be fixed by using quotes to signify hello world is a string to be printed:

````python
print("hello world")
````

#### **(P2)**

The math module provides useful functions as well as constant values such as pi. Lets use those. Write a script to find the cirumference of one end, the surface area, and volume of any cylinder using the radius and height. Then print these values. For trigonomteric functions and pi you will need to use `math.pi` or `math.cos()`. Additionally, `math.cos()` and similar trignometric functions require inputs in radians, not degrees.

**P2 Solution**

````python
radius = 1
height = 5
pi = math.pi

circumference = 2*pi*radius
surface_area = 2*pi*radius**2 + circumference*height
volume = pi*radius**2*height

print(f"The circumference of a cylinder with radius {radius} and height {height} is {circumference},\n the surface area is {surface_area},\n and the volume is {volume}")
````
