## IDE: Integrated Development Enviornment
<a id = '1a-1_IDE'></a>

Google Colab, and MATLAB are examples of integrated developement enviornments, (IDEs) an IDE is a coding enviornment which combines multiple coding activities into a single application. Within an IDE, you can write and run code and it will usually automatically identify errors that occured from running incorrect code, sometimes directing you to the specific line that contained the error. The more complex the IDE, the more you can do. JupyterLab is an example of a more complex IDE, within it you can switch between multiple programming languages as well as write in multiple different file types (.py or .ipynb), open localized terminals, get documentation help, and configure text files. If you are following this course on Google Classroom then you will make direct use of Pybricks and Google Colaboratory; but, if you downloaded this course from Github you are likely using a different IDE such as JupyterLab or Visual Studio Code (VSCode). The purpose of this notebook lesson is to teach you to use VSCode for Python and notebook files.


While there are other options when chooseing an IDE, this lesson will only show you how to do the following things to create a development environment in VSCode.

* Install Python -*This is required as it will allow you to write and run Python (.py) files in VSCode*
* Install VSCode -*This is the chosen IDE and will be where you do your work*
* Install Python Packages -*These are what allow VSCode to interact with the Python language*
* Install Jupyter Packages -*These are what allow you to open and write Jupyter Notebook files in VSCode*
* Create a Vitrual Environment -*This is a self-contained Python instance which exists in your project folder, it allows you to compartmentalize the version and packages you would like to use for a given Python project*

The PATH variable is a systemwide variable which stores the location of global commands which can be executed. When you use the command ```cd``` to change directories, or ```dir``` to list the directories you are using commands stored in the PATH variable. We would like to be able to access Python, pip, and venv commands from anywhere on the computer, therefore we should add the Python installation to the PATH. To do this we will first make a local variable:
```#newPath = "C:\Users\[user]\AppData\Local\Python\bin"```
Make sure to remember the quotation marks, otherwise you will get the error shown in the below image.

![create newpath variable](C:/Users/gerows/Python/SAIL-Plasma/PyTHAGORA/Images/newpathcorrect.png)

Next we will create a temporary variable which contains the PATH as it currently stands.
```
$currentPath = [Environment]::GetEnvironmentVariable("PATH", "User")
```
Then we can append the new path to the old path at the User level:
```
[Environment]::SetEnvironmentVariable("PATH", "$currentPath;$newPath, [EnvironmentVariableTarget]::User)
```
You can see in the below screenshot that the Python path was successfully added to the PATH variable. To make the changes we will need to restart the powershell by closing and opening again.

At this point, VSCode should recognize that you are using Python and there will now be an option in the bottom right corner to select your environment. The Python environment is the instance of Python you are using and allows VSCode (or an IDE of your choice) to compile and run Python code, it also contains all of the packages you have installed. For instance, we just installed Python 3.14.5 and have installed no packages yet. This instance is stored in the Local AppData of the User on the device you are using. This is called the "Global" Python environment. When you use the Python installation manager to update your Python version, this version is changed from an outdated version to a new one. Normally, this is the desired function; however, when writing code you sometimes become dependent on a package or function which is limited in the version of Python it can use. In this case, you may not want to update your Python version for some projects. Similarly, there may be some projects where version does not necessarily matter and you want the most recent version. To manage all of these different expectations you can create what are called "virtual environments".

A virtual environment (venv) is a snapshot of your global environment at the moment the venv is made. It will be the same version and include the same packages. What is different, is that when you update the global environment, the venv stays the same. Or, if you install packages in either environment, the other will not see those packages. This is great, as it allows for compartmentalization of your projects, limiting version errors or conflicting package installations. It is generally best practice to create a new virual environment and new directory for each separate project you work on. This prevents any weird version errors or dependency problems you could meet. 

Fortunately, making a venv is not difficult and VSCode can do it for us. In the bottom right corner look for the "Interpreter Path" button. It should be to the right of the "Language" button, which should be autoselecting Python. The "Interpreter path" button should show something like "defaultinterpreterPath" or "3.14.5".
