# Set Up

This first project is to give you an understanding of how benchmark projects like this will be formatted. Overall there will be a set-up tab which includes imports and general housekeeping before you can begin coding. Then there will be several smaller parts of increasing difficulty which will be used to test your understanding of associated material as well as related coding skills.

In this first project I will walk you through many of the solutions and how to complete the parts but as you progress through the course I will provide less and less help.
